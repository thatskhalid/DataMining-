```python
import pandas as pd
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/khali/Desktop/Coding Workspace/DataMining/data/dataset1/raw/car_prices.csv')
```

#Data Analysis Plots 

```python
#1. What Are the Most Popular Types of Vehicles?
# Count the number of vehicles sold by body type
vehicle_type_counts = df['body'].value_counts().reset_index()
vehicle_type_counts.columns = ['body_type', 'count']

# Plot the number of vehicles sold by body type
plt.figure(figsize=(12, 6))
sns.barplot(data=vehicle_type_counts, x='body_type', y='count', palette='viridis')
plt.title('Number of Vehicles Sold by Body Type')
plt.xlabel('Body Type')
plt.ylabel('Count of Vehicles Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
#2. How Many Vehicles Were Sold by Each Brand?
# Count of vehicles sold by each brand
vehicles_per_brand = df['make'].value_counts().reset_index()
vehicles_per_brand.columns = ['make', 'count']

# Plot the count of vehicles sold by each brand
plt.figure(figsize=(12, 6))
sns.barplot(data=vehicles_per_brand.head(10), x='make', y='count', palette='coolwarm')
plt.title('Top 10 Vehicle Brands by Number Sold')
plt.xlabel('Vehicle Make')
plt.ylabel('Number Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
#3. Do Different Transmission Types Affect a Vehicle's Price?
# Plot the average price by transmission type
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='transmission', y='sellingprice', palette='pastel')
plt.title('Effect of Transmission Type on Selling Price')
plt.xlabel('Transmission Type')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.show()

```

```python
#4. What Are the Most Popular Colors Being Sold?
# Count of vehicles sold by color
popular_colors = df['color'].value_counts().reset_index()
popular_colors.columns = ['color', 'count']

# Plot the most popular colors
plt.figure(figsize=(12, 6))
sns.barplot(data=popular_colors.head(10), x='color', y='count', palette='Blues')
plt.title('Top 10 Most Popular Vehicle Colors')
plt.xlabel('Color')
plt.ylabel('Number Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
#5. Is There a Correlation Between Condition and Mileage?
# Plot the correlation between condition and odometer
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='odometer', y='condition', alpha=0.6)
plt.title('Correlation Between Vehicle Condition and Mileage')
plt.xlabel('Mileage (Odometer)')
plt.ylabel('Condition')
plt.show()
```
