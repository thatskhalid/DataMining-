### Data Analysis Code // Code that answers the 30 questions located in /docs

```python
#1) What kind of new vehicles are being bought by consumers?
# Assuming "new vehicles" means vehicles with low mileage (e.g., less than 10,000 miles)
new_vehicles = df[df['odometer'] < 10000]
new_vehicle_counts = new_vehicles['make'].value_counts()
new_vehicle_counts
```

```python
#2) What can be some reasons that types of vehicles are being bought as compared to other niche types?
# This might require more detailed analysis based on body types and price.
# We'll look at body type and compare sales by body type.
body_type_sales = df['body'].value_counts()
body_type_sales
```

```python
#3) What is the current market for the following vehicle types: sedans, trucks, SUVs, compacts, hybrids?
# Filter by body type to show the market for specific types of vehicles.
vehicle_types = ['Sedan', 'Truck', 'SUV', 'Compact', 'Hybrid']
market_share = df[df['body'].isin(vehicle_types)]['body'].value_counts()
market_share
```

```python
#4) What demographics are interested in what kind of car?
# Assuming the 'state' column can represent location and thus demographics.
demographics_by_vehicle = df.groupby('state')['make'].value_counts().unstack(fill_value=0)
demographics_by_vehicle
```

```python
#5) Top selling vehicles per state or region?
# Group by 'state' and 'make' to find the top-selling vehicle per state.
top_selling_per_state = df.groupby(['state', 'make'])['sellingprice'].count().unstack(fill_value=0)
top_selling_per_state
```

```python
#6) Is there a correlation between make and condition?
# We can calculate the average condition for each make and see the correlation.
condition_by_make = df.groupby('make')['condition'].mean()
condition_by_make
```

```python
#7) Are there hybrids being sold? And is there a correlation between price versus non-hybrids?
# Assuming 'Hybrid' is a body type, filter the data.
hybrids = df[df['body'] == 'Hybrid']
non_hybrids = df[df['body'] != 'Hybrid']

# Compare the average selling prices.
hybrid_price = hybrids['sellingprice'].mean()
non_hybrid_price = non_hybrids['sellingprice'].mean()

hybrid_price, non_hybrid_price

#HYBRIDS ARE NOT A BODY TYPE 
#FLAG THIS CELL
```

```python
#8) If the data includes number of previous owners, does this affect price?
# Assuming 'previous owners' is a column.
# Group by 'previous owners' and calculate average price.
if 'previous_owners' in df.columns:
    price_by_owners = df.groupby('previous_owners')['sellingprice'].mean()
    price_by_owners
else:
    print("Previous owners data not available.")
    
#NO PREVIOUS OWNERS DATA AVAILABLE
#FLAG THIS CELL
```

```python
#9) What brands have the best resale value?
# Resale value can be calculated by comparing selling price with MMR.
resale_value = df.groupby('make').apply(lambda x: (x['sellingprice'] / x['mmr']).mean())
best_resale_value = resale_value.sort_values(ascending=False)
best_resale_value
```

```python
#10) How many vehicles were sold by each brand?
# Count the number of vehicles sold per make.
vehicles_sold_per_brand = df['make'].value_counts()
vehicles_sold_per_brand
```
