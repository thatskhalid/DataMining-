```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the cleaned data for analysis
df_cleaned = pd.read_csv('/Users/khali/Desktop/Coding Workspace/DataMining/data/cleaned/cleaned_data.csv')

# Continue with your analysis
```

```python
# 1. Distribution of Selling Prices:
# Plotting the distribution of selling prices
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['sellingprice'], bins=30, kde=True)
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Most Common Car Brands:
# Plotting the most common car brands
plt.figure(figsize=(12, 6))
df_cleaned['make'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Common Car Brands')
plt.xlabel('Car Brand')
plt.ylabel('Number of Vehicles Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Relationship Between Odometer and Selling Price:
# Scatter plot showing the relationship between odometer reading and selling price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='odometer', y='sellingprice', data=df_cleaned)
plt.title('Odometer vs Selling Price')
plt.xlabel('Odometer (miles)')
plt.ylabel('Selling Price')
plt.show()
```

```python
# 4. Condition vs Selling Price:
# Boxplot to show how vehicle condition affects selling price
plt.figure(figsize=(10, 6))
sns.boxplot(x='condition', y='sellingprice', data=df_cleaned)
plt.title('Condition vs Selling Price')
plt.xlabel('Condition')
plt.ylabel('Selling Price')
plt.show()
```

```python
# 5. Average Selling Price by Vehicle Type:
# Grouping by vehicle type (body) and plotting average selling price
plt.figure(figsize=(12, 6))
df_cleaned.groupby('body')['sellingprice'].mean().sort_values(ascending=False).plot(kind='bar', color='lightgreen')
plt.title('Average Selling Price by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45)
plt.show()
```
